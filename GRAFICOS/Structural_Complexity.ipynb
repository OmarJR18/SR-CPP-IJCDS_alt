```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import seaborn as sns
import os

os.makedirs('Structural_Complexity', exist_ok=True)

df = pd.read_csv('Data.csv')
df = df.rename(columns={'F1 score': 'F1_Score', 'Inference Time (ms) median': 'Inference_Time_ms', 'Inference Time (ms) std': 'Inference_Time_Std_ms', 'Features': 'n_Features'})

metrics = ['Nodes', 'Leaves']
methods = df['Method'].unique()
datasets = df['Dataset'].unique()

colors = ['#2F4F4F', '#007377', '#8B0000', '#B8860B', '#4B6C84', '#A0522D', '#556B2F']
markers = ['o', 's', 'D', '^', 'v', 'P', '*']

sns.set(style='whitegrid', font_scale=1.4, context='paper')
plt.rcParams['figure.dpi'] = 300

for metric in metrics:
    plt.figure(figsize=(14, 6))
    ax = plt.gca()
    for i, method in enumerate(methods):
        subset = df[df['Method'] == method].set_index('Dataset').reindex(datasets).reset_index()
        ax.plot(datasets, subset[metric], marker=markers[i % len(markers)], linestyle='-',
                label=method, color=colors[i % len(colors)], linewidth=2, markersize=8)
    ax.set_yscale('log')
    def format_func(value, tick_number):
        return f'{int(value)}' if value >= 1 else f'{value:.2f}'
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(format_func))
    ax.set_title(f'{metric} per Dataset by Method', fontsize=16, fontweight='bold')
    ax.set_xlabel('Dataset', fontsize=13, fontweight='bold')
    ax.set_ylabel(metric, fontsize=13, fontweight='bold')
    plt.xticks(rotation=45, ha='right', fontsize=11, fontweight='bold')
    plt.yticks(fontsize=11, fontweight='bold')
    ax.grid(axis='y', linestyle='--', linewidth=0.5)
    ax.legend(title='Method', fontsize=11, title_fontsize=12, loc='upper left')
    plt.tight_layout()
    plt.savefig(f'Structural_Complexity/line_{metric.lower()}.svg', format='svg', bbox_inches='tight')
    plt.close()
    print(f'Saved line_{metric.lower()}.svg')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Data.csv')
df = df.rename(columns={'F1 score': 'F1_Score', 'Inference Time (ms) median': 'Inference_Time_ms', 'Inference Time (ms) std': 'Inference_Time_Std_ms', 'Features': 'n_Features'})

methods = df['Method'].unique()
datasets = df['Dataset'].unique()

colors = ['#2F4F4F', '#007377', '#8B0000', '#B8860B', '#4B6C84', '#A0522D', '#556B2F']

plt.rcParams['font.family'] = 'serif'
plt.rcParams['figure.dpi'] = 300
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.linestyle'] = '--'
plt.rcParams['grid.linewidth'] = 0.5

fig, ax = plt.subplots(figsize=(14, 6))

for i, method in enumerate(methods):
    subset = df[df['Method'] == method].set_index('Dataset').reindex(datasets).reset_index()
    ax.plot(datasets, subset['Nodes'], marker='o', linestyle='-', linewidth=2,
            label=f'{method} - Nodes', color=colors[i % len(colors)], markersize=6)
    ax.plot(datasets, subset['Leaves'], marker='s', linestyle='--', linewidth=2,
            label=f'{method} - Leaves', color=colors[i % len(colors)], markersize=5, alpha=0.8)

ax.set_yscale('log')
ax.set_xlabel('Dataset', fontsize=13)
ax.set_ylabel('Count (log scale)', fontsize=13)
ax.set_title('Nodes and Leaves per Dataset by Method', fontsize=16, fontweight='bold')
ax.set_xticks(range(len(datasets)))
ax.set_xticklabels(datasets, rotation=45, ha='right', fontsize=11)
ax.tick_params(axis='y', labelsize=11)
ax.legend(title='Metric / Method', fontsize=10, title_fontsize=11, loc='upper left', ncol=2)
plt.tight_layout()
plt.savefig('Structural_Complexity/line_nodes_leaves.svg', format='svg', bbox_inches='tight')
plt.close()
print('Saved line_nodes_leaves.svg')
```
