```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

os.makedirs('Predictive_Performance', exist_ok=True)

df = pd.read_csv('Data.csv')
df = df.rename(columns={'F1 score': 'F1_Score', 'Inference Time (ms) median': 'Inference_Time_ms', 'Inference Time (ms) std': 'Inference_Time_Std_ms', 'Features': 'n_Features'})

datasets = df['Dataset'].unique()
methods = df['Method'].unique()

colors = [
    '#2F4F4F', '#007377', '#8B0000',
    '#B8860B', '#4B6C84', '#A0522D', '#556B2F',
]

metrics = ['Accuracy', 'F1_Score', 'Leaves', 'Nodes', 'Depth']

for metric in metrics:
    fig, ax = plt.subplots(figsize=(12, 6))
    x = np.arange(len(datasets))
    width = 0.12
    offsets = np.linspace(-width * (len(methods)-1)/2, width * (len(methods)-1)/2, len(methods))

    for i, method in enumerate(methods):
        subset = df[df['Method'] == method]
        values = subset[metric].values
        ax.bar(x + offsets[i], values, width, label=method,
               color=colors[i % len(colors)], edgecolor='black', linewidth=0.5)

    ax.set_xlabel('Dataset', fontsize=12)
    if metric in ['Leaves', 'Nodes', 'Depth']:
        ax.set_ylabel(f'{metric} (log scale)', fontsize=12)
        ax.set_yscale('log')
    else:
        ax.set_ylabel(metric.replace('_', ' '), fontsize=12)

    ax.set_title(f'{metric.replace("_", " ")} per Method and Dataset', fontsize=14, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(datasets, rotation=45, ha='right', fontsize=10)
    ax.tick_params(axis='y', labelsize=10)

    legend = ax.legend(title='Method', fontsize=10, title_fontsize=12,
                       loc='center left', bbox_to_anchor=(1, 0.5), frameon=True)
    frame = legend.get_frame()
    frame.set_edgecolor('black')
    frame.set_alpha(0.9)
    frame.set_linewidth(0.8)
    frame.set_facecolor('white')

    ax.yaxis.grid(True, linestyle='--', linewidth=0.5, alpha=0.6)
    ax.set_axisbelow(True)
    plt.tight_layout(rect=[0, 0, 0.85, 1])
    plt.savefig(f'Predictive_Performance/bar_{metric.lower()}.svg', format='svg', bbox_inches='tight')
    plt.close()
    print(f'Saved bar_{metric.lower()}.svg')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('Data.csv')
df = df.rename(columns={'F1 score': 'F1_Score', 'Inference Time (ms) median': 'Inference_Time_ms', 'Inference Time (ms) std': 'Inference_Time_Std_ms', 'Features': 'n_Features'})

datasets = df['Dataset'].unique()
methods = df['Method'].unique()

colors = [
    '#2F4F4F', '#007377', '#8B0000',
    '#B8860B', '#4B6C84', '#A0522D', '#556B2F',
]

metrics = ['Accuracy', 'F1_Score', 'Leaves', 'Nodes', 'Depth', 'Inference_Time_ms']

for metric in metrics:
    fig, ax = plt.subplots(figsize=(12, 6))
    x = np.arange(len(datasets))
    width = 0.12
    offsets = np.linspace(-width * (len(methods) - 1) / 2, width * (len(methods) - 1) / 2, len(methods))

    for i, method in enumerate(methods):
        subset = df[df['Method'] == method]
        if metric == 'Inference_Time_ms':
            means = subset[metric].values
            errors = subset['Inference_Time_Std_ms'].values
            ax.bar(x + offsets[i], means, width, label=method, yerr=errors, capsize=4,
                   color=colors[i % len(colors)], edgecolor='black', linewidth=0.5)
        else:
            values = subset[metric].values
            ax.bar(x + offsets[i], values, width, label=method,
                   color=colors[i % len(colors)], edgecolor='black', linewidth=0.5)

    ax.set_xlabel('Dataset', fontsize=12)
    if metric in ['Leaves', 'Nodes', 'Depth']:
        ax.set_ylabel(f'{metric.replace("_", " ")} (log scale)', fontsize=12)
        ax.set_yscale('log')
    else:
        ax.set_ylabel(metric.replace('_', ' '), fontsize=12)

    ax.set_title(f'{metric.replace("_", " ")} per Method and Dataset', fontsize=14, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(datasets, rotation=45, ha='right', fontsize=10)
    ax.tick_params(axis='y', labelsize=10)
    ax.yaxis.grid(True, linestyle='--', linewidth=0.5, alpha=0.6)
    ax.set_axisbelow(True)

    legend = ax.legend(title='Method', fontsize=10, title_fontsize=11,
                       loc='upper center', bbox_to_anchor=(0.5, -0.40), ncol=len(methods), frameon=True)
    frame = legend.get_frame()
    frame.set_edgecolor('black')
    frame.set_alpha(0.95)
    frame.set_linewidth(0.8)
    frame.set_facecolor('white')

    fig.subplots_adjust(bottom=0.3)
    plt.savefig(f'Predictive_Performance/bar_with_error_{metric.lower()}.svg', format='svg', bbox_inches='tight')
    plt.close()
    print(f'Saved bar_with_error_{metric.lower()}.svg')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('Data.csv')
df = df.rename(columns={'F1 score': 'F1_Score', 'Inference Time (ms) median': 'Inference_Time_ms', 'Inference Time (ms) std': 'Inference_Time_Std_ms', 'Features': 'n_Features'})

colors = [
    '#2F4F4F', '#007377', '#8B0000',
    '#B8860B', '#4B6C84', '#A0522D', '#556B2F',
]

sns.set(style='whitegrid', font_scale=1.4, context='paper')
plt.rcParams['figure.dpi'] = 300

for metric in ['Accuracy', 'F1_Score']:
    plt.figure(figsize=(12, 6))
    ax = sns.boxplot(data=df, x='Method', y=metric, palette=colors)
    ax.set_title(f'{metric.replace("_", " ")} per Method', fontsize=16, fontweight='bold')
    ax.set_xlabel('Method', fontsize=13, fontweight='bold')
    ax.set_ylabel(metric.replace('_', ' '), fontsize=13, fontweight='bold')
    plt.xticks(rotation=45, ha='right', fontsize=11, fontweight='bold')
    plt.yticks(fontsize=11, fontweight='bold')
    plt.grid(axis='y', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.savefig(f'Predictive_Performance/box_{metric.lower()}.svg', format='svg', bbox_inches='tight')
    plt.close()
    print(f'Saved box_{metric.lower()}.svg')
```
